3. Count occurrences of each word in a sentence

```python
sentence = "Hii welcome to assongment, most welcome"
words = sentence.split()
word_count = {}

for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print(word_count)
```

7. Greatest Common Divisor (GCD)

```python
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

print(gcd(20, 28))
print(gcd(98, 56))
```

8. Check whether a list contains a sublist

```python
main_list = [1, 2, 3, 4, 5, 6, 7, 8]
sub_list = [3, 4, 5]

exists = False
for i in range(len(main_list)):
    if main_list[i:i+len(sub_list)] == sub_list:
        exists = True

print("Sublist exists:", exists)
```

13. Sort a dictionary by value

```python
d = {'a': 3, 'b': 1, 'c': 2}

asc = dict(sorted(d.items(), key=lambda x: x[1]))
desc = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))

print("Ascending:", asc)
print("Descending:", desc)
```

15. Fibonacci series up to n terms

```python
n = int(input("Enter the number:"))
fib = [0, 1]

for i in range(2, n+1):
    fib.append(fib[i-1] + fib[i-2])

print(fib)
```

17. Sum of odd and even series using function

```python
import math

def odd_series(n):
    total = 0
    for i in range(1, n+1, 2):
        total += (i**2) / math.factorial(i)
    return total

def even_series(n):
    total = 0
    for i in range(2, n+1, 2):
        total += (i**2) / math.factorial(i)
    return total

print("Odd series sum:", odd_series(10))
print("Even series sum:", even_series(10))
```

18. Factorial using recursion

```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(10))
```
